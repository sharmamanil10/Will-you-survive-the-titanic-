```python
def f(s):
    if s=='male':
        return 0
    else:
        return 1
def g(s):
    if s=='S':
        return 0
    else:
        return 1
    
    
```

```python
import numpy as np
import pandas as pd
titanic_train=pd.read_csv("training_titanic_x_y_train.csv",delimiter=",")
titanic_test=pd.read_csv("test_titanic_x_test.csv",delimiter=",")
df2=titanic_test.copy()
df=titanic_train.copy()
df.head()
df.drop('Name',axis=1,inplace=True)
df.drop('Cabin',axis=1,inplace=True)
df=df.dropna()
df.reset_index(drop=True,inplace=True)
target=df.Survived
df.drop('Survived',axis=1,inplace=True)
df.drop('Ticket',axis=1,inplace=True)
df["Sex"]=df.Sex.apply(f)
df["Embarked"]=df.Embarked.apply(g)
#extras
df["age_age"]=df['Age']*df['Age']
df["pclass_pclass"]=df['Pclass']*df['Pclass']
df["Pclass_Age"]=df['Pclass']*df['Age']
df["Embarked_pclass"]=df['Embarked']*df['Pclass']
df["Embarked_pclass_age"]=df['Embarked']*df['Pclass']*df['Age']
print(df.head())


df2.drop('Name',axis=1,inplace=True)
df2.drop('Cabin',axis=1,inplace=True)

df2["Sex"]=df2.Sex.apply(f)
df2.reset_index(drop=True,inplace=True)
df2.drop('Ticket',axis=1,inplace=True)
df2["Embarked"]=df.Embarked.apply(g)
df2.Age.fillna(df2.Age.mean(),inplace=True)
df2["age_age"]=df2['Age']*df2['Age']
df2["pclass_pclass"]=df2['Pclass']*df2['Pclass']
df2["Pclass_Age"]=df2['Pclass']*df2['Age']
df2["Embarked_pclass"]=df2['Embarked']*df2['Pclass']
df2["Embarked_pclass_age"]=df2['Embarked']*df2['Pclass']*df2['Age']
#target=df2.Survived
#df.drop('Survived',amean(),inplace=True)
df2.describe()
```
